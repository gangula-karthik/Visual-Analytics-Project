```
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_primary_schools_in_Singapore"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find("table", {"class": "wikitable"})
```

```
table_data = []
for row in table.find_all('tr'):
    cols = row.find_all('td')
    cols = [col.text.strip() for col in cols]
    table_data.append(cols)

table_data
```

```
import pandas as pd

# Define the column names for the DataFrame
columns = ['Primary_school', 'Funding', 'Type', 'Area', 'Notes', 'Column 6', 'Column 7']

# Create the DataFrame with the extracted data and column names
df = pd.DataFrame(table_data[1:], columns=columns)
# remove last 3 columns
df = df.drop(df.columns[-3:], axis=1)

df
```

```
import requests
import json

# iterate over each row in the DataFrame
for index, row in df.iterrows():
    address = row['Primary_school']
    url = f'https://developers.onemap.sg/commonapi/search?searchVal={address}&returnGeom=Y&getAddrDetails=Y&pageNum=1'
    response = requests.get(url)
    data = json.loads(response.text)
    if data['found'] > 0:
        lat = data['results'][0]['LATITUDE']
        lon = data['results'][0]['LONGITUDE']
        df.at[index, 'LATITUDE'] = lat
        df.at[index, 'LONGITUDE'] = lon
    else:
        df.at[index, 'LATITUDE'] = 'Not found'
        df.at[index, 'LONGITUDE'] = 'Not found'


# drop rows where latitude or longitude is "Not Found"
df = df[(df.LATITUDE != 'Not found') & (df.LONGITUDE != 'Not found')]

df
```

```
import re

# remove the square brackets from the "Mall Name" column
df_malls["Mall Name"] = df_malls["Mall Name"].str.replace(r"\[.*?\]", "")

df_malls
```

```

import requests

# create new columns for latitude and longitude
df_malls["LATITUDE"] = None
df_malls["LONGITUDE"] = None

# iterate over each mall in the "Mall Name" column
for index, row in df_malls.iterrows():
    mall_name = row["Mall Name"]
    # construct the API url with the mall name
    url = f"https://developers.onemap.sg/commonapi/search?searchVal={mall_name}&returnGeom=Y&getAddrDetails=N&pageNum=1"
    # make the GET request
    response = requests.get(url)
    # check if the request was successful
    if response.status_code == 200:
        # retrieve the latitude and longitude from the response
        data = response.json()
        try:
            if 'results' in data and len(data['results']) > 0:
                latitude = data['results'][0]['LATITUDE']
                longitude = data['results'][0]['LONGITUDE']
                # add the latitude and longitude to the DataFrame
                df_malls.at[index, "LATITUDE"] = latitude
                df_malls.at[index, "LONGITUDE"] = longitude
            else:
                print(f"No results found for {mall_name}")
        except IndexError:
            print(f"An error occurred while trying to retrieve location for {mall_name}")

print(df_malls)
```

```
import pandas as pd
from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371 # Radius of earth in kilometers.
    return c * r

for index, row in housing_dist.iterrows():
    min_distance = float('inf')
    min_school = None
    for i, r in schools_coord.iterrows():
        distance = haversine(row['longitude'], row['latitude'], r['LONGITUDE'], r['LATITUDE'])
        if distance < min_distance:
            min_distance = distance
            min_school = r['Primary_school']
    housing_dist.at[index, 'nearest_school'] = min_school
    housing_dist.at[index, 'distance_to_school'] = min_distance

```
