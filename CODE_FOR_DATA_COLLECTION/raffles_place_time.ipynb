```
main_url = "https://developers.onemap.sg/privateapi/routingsvc/route?"
time_taken = []


for index, rows in mrt_stations.iterrows(): 
    lat = rows['lat']
    lon = rows['lng']
    temp = main_url + "start=" + str(lat) + "," + str(lon) + "&" + "end=1.284001,103.85155" + f"&routeType={params['routeType']}&" + f"token={token}&" + f"date={params['date']}&" + f"time={params['time']}&" + f"mode={params['mode']}"
    print(temp)
    if lat == 1.284001 and lon == 103.85155: 
        time_taken.append(0)
        continue
    req = requests.get(temp)
    if req.status_code == 404: 
        print(lat, lon, "has an error")
        time_taken.append(None)
        continue
    ans = json.loads(req.text)["plan"]["itineraries"][0]["transitTime"] / 60
    time_taken.append(ans)
```

```
temp_df = pd.DataFrame(time_taken)
```

```
mrt_stations = mrt_stations.join(temp_df)

mrt_stations
```

```
# Get the name of the last column in the DataFrame
last_column = mrt_stations.columns[-1]

# Rename the last column
mrt_stations.rename(columns={last_column: 'time_taken_to_raffles_place_mrt'}, inplace=True)
```
