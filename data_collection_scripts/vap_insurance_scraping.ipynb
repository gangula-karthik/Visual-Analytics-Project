```python
import tabula
import pandas as pd

# read pdf into a DataFrame
df = tabula.read_pdf("/content/v8-comparison-of-std-ips---sep-2022.pdf", pages=3)

df
```

```python
headers = ['Age Next Birthday:','IncomeShield','HealthShield Gold Max','GREAT SupremeHealth','PRUShield Standard','Singlife Shield','AXA Shield	','Raffles Shield']
data_frame = pd.concat(df,keys=headers, axis=1)
data_frame = data_frame.drop(data_frame.index[:3])
data_frame = data_frame.reset_index(level=0, drop=True)

data_frame
```
