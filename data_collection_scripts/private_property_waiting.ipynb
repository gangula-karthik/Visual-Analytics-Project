```
import pandas as pd
import requests
from bs4 import BeautifulSoup


def getter_top_dates(): 
    property_names = []
    top_date = []

    for i in range(1, 10): 
        url = f"https://esingaporeproperty.sg/property-type/condo-private-new/page/{i}/"
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")

        ul_element = soup.find("ul", class_="large-block-grid-2 medium-block-grid-2")
        
        for div_element in ul_element.find_all("div", class_="img-wrap"):
            div_element.decompose()

        li_elements = str(ul_element).split("<div class=\"card\">")[1:]
        for i in li_elements:
            soup = BeautifulSoup(i, "html.parser")
            prop_element = soup.find("div", class_="title")
            print(prop_element)

            top_element = soup.find_all("li") 
            third_li = top_element[2]
            print(str(third_li.text))

            if top_element is None: 
                top_date.append(None)
            else: 
                top_date.append(str(third_li.text))
        
            text = prop_element.a.text
            property_names.append(text)

    return pd.DataFrame(data={'Property_names': property_names, 'TOP_year': top_date})


getter_top_dates()
```
